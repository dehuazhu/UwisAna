First identify the encoding of the .csv file

```python
import chardet
rawdata = open('data.csv','rb').read()
result  = chardet.detect(rawdata)
charenc = result['encoding']
print(charenc)
```

Now import the data into a pandas dataframe

```python
import pandas as pd
data = pd.read_csv('data.csv',encoding=charenc)
```

Directly from our dataframe, we can plot the distributions for all of our features, just to see how everything looks

```python
import matplotlib
data.hist(bins=20, figsize=(10,10),color='black');
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,10))
sns.heatmap(data.corr())
```

```python
# get the ULTIMATE OVERSIGHT (overkill) of the DATAAAAAAAAAAAAAAAAAAAAAAAAAAA!!!!
sns.pairplot(data)
```
